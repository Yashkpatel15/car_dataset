```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("quikr_car.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['kms_driven']=df['kms_driven'].str.replace('kms','')
df['kms_driven']=df['kms_driven'].str.replace(',','')
df['kms_driven']=df['kms_driven'].str.replace('Petrol','')
df['kms_driven']=pd.to_numeric(df['kms_driven'],errors='coerce')
```

```python
df['kms_driven']=df['kms_driven'].fillna(method='ffill').fillna(method='bfill')
```

```python

df['fuel_type']=df['fuel_type'].fillna(method='ffill').fillna(method='bfill')
```

```python
df['Price']=df['Price'].str.replace(',','')
df['Price']=df['Price'].str.replace('Ask For Price','')
df['Price']=pd.to_numeric(df['Price'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['Price']=df['Price'].fillna(df['Price'].median())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['year'].unique()
```

```python
df['year']=pd.to_numeric(df['year'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['year']=df['year'].fillna(method='ffill').fillna(method='bfill')
```

```python
df['Price']=df['Price'].astype(int)
df['kms_driven']=df['kms_driven'].astype(int)
df['year']=df['year'].astype(int)
```
